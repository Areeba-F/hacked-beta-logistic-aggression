**Graph showing all correlations**


*   Creates dataframe containing all the correlation values between each data point
*   Creates heatmap representation of correlation values



```python
corr = df.corr()
sns.set(rc = {'figure.figsize':(13,10)})
sns.heatmap(corr, annot=True, fmt = ".0%")
```

**Graph showing correlation between Cardio disease**


*   Creates dataframe containing all the correlation values between each data point
*   Splits data to only contain the row relevant to cardio disease (and removes correlation between cardio disease and itself)
*   Creates heatmap representation of correlation values






```python
corr2 = df.corr()
corr2 = corr2.iloc[0:11,11:].T
sns.set(rc = {'figure.figsize':(20,2)})
sns.heatmap(corr2, vmin=-.08, vmax=1, annot=True, fmt = ".0%")
```

**Graph showing age range of patients with cardio**

*   Creates dataframe containing the values of 'age(days)' column where the 'cardio' column values are 1
*   Divides all values in dataframe by 356 to get years
*   Creates distribution plot representation of age values


```python
age_df = df.loc[df['cardio'] == 1, 'age(days)']
age_df = age_df/365
sns.set(rc = {'figure.figsize':(12,10)})
sns.distplot(age_df, bins = 10)
```
